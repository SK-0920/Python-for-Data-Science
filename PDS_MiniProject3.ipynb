3.	Find which country's players data is maximum in the dataset and what’s is the count of it (2 Marks)

```python
#Players count acc. to Nationality
country_counts = fifa["Nationality"].value_counts()
print(country_counts.head())
```

```python
# Country & its Count
top_country = country_counts.idxmax()
top_count = country_counts.max()
print("Country with highest number of players:", top_country)
print("Players count:", top_count)
```

4.	Display top 10 clubs ordered by sum of all player’s current market value of that club (3 Marks)

```python
#Each Club value
club_values = fifa.groupby("Club")["Value"].sum()
print(club_values.sort_values(ascending=False)) 
```

```python
#Top 10 clubs by total player market value
top10_clubs = club_values.sort_values(ascending=False).head(10).reset_index()
print(top10_clubs)
```

5.	Count the number of players having International Reputation ratings = 4 from the clubs 'Real Madrid’ and ‘FC Barcelona’ (3 Marks)

```python
#Number of players in both clubs having International Reputation ratings = 4
count_players = fifa[
            (fifa["International Reputation"] == 4) & (fifa["Club"].isin(["Real Madrid", "FC Barcelona"]))].shape[0]
print("Number of players in both clubs having International Reputation ratings = 4 is:", count_players)
```

```python
#Name of players in both clubs having International Reputation ratings = 4
players_list = fifa[
    (fifa["International Reputation"] == 4) &
    (fifa["Club"].isin(["Real Madrid", "FC Barcelona"]))][["Name", "Club"]]

print("Players with IR = 4 from Real Madrid & FC Barcelona:\n")

for club, group in players_list.groupby("Club"):
    print(f"{club}")
    for name in group["Name"]:
        print(f"-{name}")
    print()
```

```python
#Count of players with IR=4 from each clubs
club_wise_count = fifa[
    (fifa["International Reputation"] == 4) &
    (fifa["Club"].isin(["Real Madrid", "FC Barcelona"]))].groupby("Club").size().to_frame("Player_Count").reset_index()

print("Count of players with IR=4 from each clubs\n", club_wise_count)
```

6.	Find out the names of the youngest and oldest players from the dataset (5 Marks)

```python
# Youngest players
youngest_players = fifa[fifa["Age"] == fifa["Age"].min()][["Name", "Age", "Club"]].reset_index(drop=True)

# Oldest players
oldest_players = fifa[fifa["Age"] == fifa["Age"].max()][["Name", "Age", "Club"]].reset_index(drop=True)

#Name List of youngest & oldest players
print("Youngest Players:\n", youngest_players)
print()
print("Oldest players:\n", oldest_players)
```

5.	Display the top 5 countries who won the most medals. Support your answer with suitable 

```python
#reference
display(merged.head())
```

```python
#Countries with most medals
most_medals = (
    athlete.groupby("Team")["Medal"]
    .count()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={"Team": "Country", "Medal": "Total_Medals"})
)

print(most_medals)          
```

```python
#Top 5 countries
top_countries = most_medals.head(5)
display(top_countries)
```

```python
#Plot
plt.figure(figsize=(8, 5))
sns.barplot(x="Country",
            y="Total_Medals",
            hue="Country",
            data=top_countries,
            palette="viridis",
            legend=False
           )
plt.xlabel("Country")
plt.ylabel("Number of Medals")

#Show values on top of bar
for i, val in enumerate(top_countries["Total_Medals"]):
    plt.text(i, val+50, str(val), ha="center", va="bottom")


plt.show()
```
